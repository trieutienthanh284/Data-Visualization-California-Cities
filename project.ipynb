```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/california_cities.csv')
df.head()
```

```python
df = df.drop(columns = ['area_total_sq_mi', 'area_land_sq_mi', 'area_water_sq_mi', 'elevation_ft'])
df.head()
```

```python
lat, longd = df['latd'], df['longd']
population = df['population_total']
area = df['area_total_km2']
plt.scatter(longd, lat, c = np.log10(population), cmap = 'viridis',
            s = area, linewidths = 0, alpha = 0.8)
plt.style.use('seaborn-v0_8-darkgrid')
plt.colorbar(label = 'log$_{10}$(population)');
plt.axis('equal');
plt.xlabel('Latitude');
plt.ylabel('Longitude');
plt.clim(2,7)

area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([], [], s = area,
                label = str(area) + 'km$^2$')
plt.legend(labelspacing = 1, title = 'Area');
plt.title('Population and Area Distribtion of California City')
```
